```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/home/jorgebdelafuente/Doctorado/Quantics/PH2p_1TApp/ph2p_bckup/ph2p_v14"
flux_rg = np.loadtxt(path + "/flux_rg",skiprows=5)

fig, ax = plt.subplots(1,1)
ax.plot(flux_rg[:,0],flux_rg[:,3])
plt.show()
```

```python
amutokg = 1.660539040E-27

massp = 30.9737620*amutokg
massd = 2.0141017778*amutokg

rmd2 = (massd*massd)/(massd+massd)
rmpd2 = (rmd2*massp)/(rmd2+massp)

eVtoJ = 1.602176634E-19
hbar = 1.054571800E-34

# rmpd2 = rmpd2*amutokg

sigma_J = np.zeros(len(flux_rg[:,3]))
for i,e,prob in zip(range(len(flux_rg[:,3])),flux_rg[:,0],flux_rg[:,3]):
    e = e*eVtoJ
    k2 = (2*rmpd2*e)/(hbar*hbar)
    sigma_J[i] = (np.pi/k2)*prob

fig, ax = plt.subplots(1,1)
ax.plot(flux_rg[:,0],sigma_J)
# ax.set_ylim(-0.2,1000000)
plt.show()
```
